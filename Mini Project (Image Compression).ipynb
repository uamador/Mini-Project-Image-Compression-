# 2 - Reading an image
Read an image using matplotlib.  
Hint: You can use any alternative library for reading images. Check out OpenCV, skimage, etc.

```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.image as mpimg
img = mpimg.imread('cat.png')
img.shape
```

# 3 - Re-organize
Re-organize the pixels as a 2D array with dimensions (number of pixels, 3).  
In other words, a dataset with number of pixels samples and 3 columns.  

```python
pixel_data = img.reshape(-1, img.shape[-1])
pixel_data.shape
```

# 4 -  Clustering
Use K-means to cluster the data into 10 clusters.  
Replace each pixel by its cluster center.  
Re-organize the pixels into a 3D array to show the image.  

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

k = 10

kmeans = KMeans(n_clusters=k)
kmeans.fit(pixel_data)

# Replace each pixel value with its corresponding centroid
compressed_data = kmeans.cluster_centers_[kmeans.labels_]


# Reshape the compressed pixel data to its original shape
compressed_img = compressed_data.reshape(img.shape)

# Show the image
plt.imshow(compressed_img)
plt.show()
```

### Question:  
What happens to the compressed image if k increases?  
A: A higher k results in less compression and a more accurate image. The higher the k the longer it takes to render the image.

# 5 - Save the image
Save the image as compressed.png.  
Is there a difference in the sizes of the two images?

```python
mpimg.imsave('compressed.png', compressed_img)
```

Yes, there is significant size difference. The original is about 650k, while the compressed one is about 140k.
